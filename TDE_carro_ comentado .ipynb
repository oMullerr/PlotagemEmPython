## Curvas de Bézier

```python
# importando as bibliotecas necessarias 
from matplotlib.path import Path
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# define the geometry dash of a path:
verts1 = [(3.8, 0.8),(3.8,2.2),(5,2.2),(6.2, 2.2), (6.2, 0.8)] # roda 1
verts2 = [(8.8, 0.8),(8.8,2.2), (10,2.2),(11.2, 2.2),(11.2, 0.8)] # roda 2
Roda1Contorno = [(3.7, 0.9),(3.7,2.3),(5,2.3),(6.3, 2.3), (6.3, 0.9)] # roda 1 contorno
Roda2Contorno = [(8.7, 0.9),(8.7,2.3), (10,2.3),(11.2, 2.3),(11.3, 0.9)] # roda 2 contorno
lanternafrente = [(3.1, 3.5),(3.1, 2.5),(2.3, 2.5)] # Farol dianteiro
lanternatras = [(11.7, 3.3),(10.8, 2.8),(12, 2.3)] # lanternas traseira
verts5 = [(3.1, 3.5), (3.1, 3.5),(4.5, 3.5)] # capô
frente = [(2.3, 2),(2.3, 3.3),(3.1, 3.5)]
cimaParabrisa = [(5.6, 4.8), (5.7, 4.5), (6, 4.5)] # Parte de cima do parabrisa
verts6 = [(5.5, 4.6),(5.5 , 4.8),(6.1 ,4.8)]
verts7 = [(11.5, 3.5),(12, 3.5),(12, 2)]
verts8 = [(12.3, 1.3),(12.3, 0.8),(11.2, 0.8)]

verts8_2 = [(12.3, 1.4),(12.3, 0.9),(11.2, 0.9)]
curvaPorta1 = [(5.9,2.4),(6.6, 2.2),(6.6 ,0.9)] # curva da porta
baixoParabrisa = [(4.5,3.5),(4.7, 3.2),(5.5, 3.2)] # parte de baixo do parabrisa
baixoCarro = [(2.3, 1.3),(2.3, 0.8),(3.8, 0.8)] # baixa da roda 1
baixoCarro2 = [(2.3, 1.4),(2.3, 0.9),(3.8, 0.9)] # baixa da roda 1
lanternafim = [(2.6, 1.2),(2.9,1.5),(2.6,1.8)] # lanterna
portacurvafim = [(8.45, 3.2),(9,2.8),(8.45,0.9)] # curva da porta

codes = [Path.MOVETO, # put the "pen" at P0
         Path.CURVE3, # and draw bezier curves
         Path.CURVE3, # from P0 to P1, to P2
         ]
codes2 = [Path.MOVETO, # put the "pen" at P0
         Path.CURVE3, # and draw bezier curves
         Path.CURVE3,
          Path.CURVE3,
          Path.CURVE3# from P0 to P1, to P2
         ]
path = Path(verts1, codes2) # Definimos o caminho do path
path2 = Path(verts2, codes2) # Definimos o caminho do path2
path5 = Path(verts5, codes) # Definimos o caminho do path5
path6 = Path(verts6, codes) # Definimos o caminho do path6
path7 = Path(verts7, codes) # Definimos o caminho do path7
path8 = Path(verts8, codes) # Definimos o caminho do path8
path8_2 = Path(verts8_2, codes) # Definimos o caminho do path8_2
pathfrente = Path(frente, codes) # Definimos o caminho do pathfrente
pathcima1 = Path(cimaParabrisa, codes) # Definimos o caminho do pathcima1
pathbaixoPara = Path(baixoParabrisa, codes) # Definimos o caminho do pathbaixoPara
pathContornoRoda1 = Path(Roda1Contorno, codes2) # Definimos o caminho do pathContornoRoda
pathContornoRoda2 = Path(Roda2Contorno, codes2) # Definimos o caminho do pathContornoroda2
pathbaixo = Path(baixoCarro, codes) # Definimos o caminho do pathbaixo
pathbaixo2 = Path(baixoCarro2, codes) # Definimos o caminho do pathbaixo2
pathPorta1 = Path(curvaPorta1, codes) # Definimos o caminho do pathporta1
pathLanterna = Path(lanternafrente, codes) # Definimos o caminho do pathLanterna
pathLanterna2 = Path(lanternatras, codes) # Definimos o caminho do pathLanterna2
pathlanternadianteira = Path(lanternafim, codes) # Definimos o caminho do pathlanternadianteira
pathportacurva = Path(portacurvafim, codes) # Definimos o caminho do pathportacurva
```

```python
def reta(p0,p1, color="black", linewidth="2"): # função das retas
    #equação da reta
    U = [x for x in range(2)]
    vx = [p0[0] + u*(p1[0]-p0[0]) for u in U]
    vy = [p0[1] + u*(p1[1]-p0[1]) for u in U]
    plt.plot(vx,vy, color=color)
    
#---------------------------------------------------------------------------------------------#   

def retacontorno(p0,p1, color="black", linewidth="3"): # função dos contornos da reta
    U = [x for x in range(2)]
    vx = [p0[0] + u*(p1[0]-p0[0]) for u in U]
    vy = [p0[1] + u*(p1[1]-p0[1]) for u in U]
    plt.plot(vx,vy, color=color)
```
